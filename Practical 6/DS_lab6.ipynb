```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
print(df.head())

```

```python
print(df.info())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), kde=True, bins=30, color="skyblue")
plt.title("Distribution of Age")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df, palette="Set2")
plt.title("Passenger Survival Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Fare'], color="lightgreen")
plt.title("Fare Distribution (Boxplot)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data=df, x="Age", hue="Survived", bins=30, kde=True, palette="coolwarm")
plt.title("Age Distribution by Survival")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", hue="Survived", data=df, palette="pastel")
plt.title("Survival Count by Gender")
plt.show()


```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="Survived", y="Fare", data=df, palette="muted")
plt.title("Fare vs Survival")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Pclass", hue="Survived", data=df, palette="husl")
plt.title("Survival Count by Passenger Class")
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

data = [12, 15, 14, 10, 18, 20, 22, 21, 19, 17]

s = pd.Series(data)

mean_val = s.mean()
median_val = s.median()
std_val = s.std()
var_val = s.var()
skew_val = skew(s)
kurt_val = kurtosis(s)

print("Mean:", mean_val)
print("Median:", median_val)
print("Standard Deviation:", std_val)
print("Variance:", var_val)
print("Skewness:", skew_val)
print("Kurtosis:", kurt_val)
```
